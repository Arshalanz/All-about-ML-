#matplotlib use to plot the Values

```python
#import numpy to get value
import numpy  as np
```

```python
#import matplotlib to plot the values
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(y)
```

```python
#plotting tha data value
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title xlabel and y label and angel
plt.plot(x,y)
plt.title("sine wave")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#plot parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r+")
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), "g*")
plt.plot(x,np.cos(x),"r-")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'French', 'German', 'Italian', 'Japanese', 'Chinese']
people = [10,8,6,4,2,2,1]
ax.bar(languages,people)
plt.xlabel("Languages")
plt.ylabel("People")
plt.title("Popular Languages")
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'French', 'German', 'Italian']
people = [100,50,150,40,70]
ax.pie(people,labels = languages, autopct='%1.1f%%')
plt.show()
```

```python
x= np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'r')
ax.scatter(x,z,color = 'b')

plt.show()
```

#3D scatter plot

```python
fig3 =  plt.figure()
ax = plt.axes(projection = '3d')
z = 20 *np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap = 'Blues')
plt.show()
```
