#Label Encoding:
. Convert the labels innto numeric form


```python
#importing the dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

label encoding of breast cancer dataset

```python
#loading the data from  csv file to pandas dataframe
cancer_data = pd.read_csv('/content/breast_cancer_data (1).csv')
```

```python
#first five reow s of the dataframe
cancer_data.head()
```

```python
cancer_data.shape
```

```python
#finding the counts of Diffrent labels
cancer_data['diagnosis'].value_counts()
```

```python
#Load the Label encoder function
label_encode = LabelEncoder()
```

```python
labels  = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#Appending the labels to The DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0. ---> BENIGN
1. ---> MALIGANT

```python
cancer_data['target'].value_counts()
```
