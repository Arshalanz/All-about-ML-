#imbalanced Dataset
A dataset with an unequal class distribution

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv("/content/credit_data.csv")
```

```python
#first five rows of the dataFrame
credit_card_data.head()
```

```python
credit_card_data["Class"].value_counts()
```

```python
#imbalenced dataset
# 0 ---> legit transaction
# 1 ---> fraud transaction
```

```python
#Seprating the legit and fraudlent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling


```python
#Build a sample dataset containing similar distribution of legit transaction and fraudlent transaction
```

Number of Fraudlent transaction ---> 38





```python
legit_sample = legit.sample(n = 500)
```

```python
legit_sample.shape
```

Conatenate two dataset

```python
new_dataset =  pd.concat([legit_sample, fraud], axis = 0)
```
